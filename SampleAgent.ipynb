```python
from google.adk.agents import Agent
from google.adk.tools import google_search
from google.adk.runners import InMemoryRunner
from google.genai import types
```

```python
#Define an agent with Google Search tool
root_agent = Agent(
    name = "Search_agent",
    description="A simple agent that can answer general questions",
    model = "gemini-2.5-flash-lite",
    tools = [google_search],
    instruction="You are a helpful assistant who answers in less than 5 words.Use google search for current information or if unsure",
)
```

```python
runner = InMemoryRunner(agent=root_agent)
```

```python
response = await runner.run_debug("What problems plague the development of India?")
```
