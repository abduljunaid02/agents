```python
#Imports 

from google.adk.agents import Agent, SequentialAgent, LoopAgent, ParallelAgent
from google.adk.tools import google_search, FunctionTool, AgentTool
from google.adk.runners import InMemoryRunner
from google.genai import types
```

```python
#Agent no. 1 - The Researcher

research_agent = Agent(
    name = "researchAgent",
    description="An agent that researches on a particular topic",
    model = "gemini-2.5-flash-lite",
    tools = [google_search],
    instruction = "You are a top research assistant. " \
    "You will use google_search tool to deep dive into a topic and" \
    " give 2-3 relevant pieces of information from different sources",
    output_key="research_findings"
)
```

```python
#Agent no. 2 - The Summariser


summary_agent = Agent(
    name = "summaryAgent",
    description="An agent that summarises a particular topic",
    model = "gemini-2.5-flash-lite",
    instruction = "You are a very good editor who has great summarisation skills. " \
        "You will take the {research_findings} and summarise them into a concise summary of max 50 words.",
    output_key="summary_output"
)
```

```python
root_agent = Agent(
    name = "Research_and_Summarise_Agent",
    description = "An agent to coordinate research and summarisation",
    model = "gemini-2.5-flash-lite",
    tools = [AgentTool(research_agent),
             AgentTool(summary_agent)],
    instruction="""You are a research coordinator. Your goal is to answer the user's query by orchestrating a workflow.
1. First, you MUST call the `ResearchAgent` tool to find relevant information on the topic provided by the user.
2. Next, after receiving the research findings, you MUST call the `SummarizerAgent` tool to create a concise summary.
3. Finally, present the final summary clearly to the user as your response."""
)
```

```python
runner = InMemoryRunner(root_agent)
response = await runner.run_debug("What is the state of Indian politics with respect to the culture of welfare schemes?")
```
